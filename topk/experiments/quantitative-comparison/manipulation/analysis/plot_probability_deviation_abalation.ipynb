```python
import numpy as np
import os

folder_path='experiments/quantitative-comparison/manipulation/analysis/data/Ablation' # path to your folder
loaded_data = {}

for filename in os.listdir(folder_path):
    if filename.endswith('.npz'):
        file_path = os.path.join(folder_path, filename)
        data = np.load(file_path)
        variable_name = filename.split('.')[0]
        loaded_data[variable_name] = [data[f'arr_{i}'] for i in range(len(data.files))]
        print(f"Loaded {variable_name}")
```

```python
loaded_data["vit-b-16-top5-d12-p100"][1].mean()
```

```python
import pandas as pd
import re

rows = []

for k, v in loaded_data.items():
    match = re.match(r'vit-b-(\d+)-top(\d+)-d(\d+)-p(\d+)', k)
    if match:
        patch_size = int(match.group(1))
        top_k = int(match.group(2))
        depth = int(match.group(3))
        mask = int(match.group(4))
        acc = v[0].mean()
        
        rows.append([patch_size, top_k, depth, mask, acc])

df = pd.DataFrame(rows, columns=["patchSize", "topK", "depth", "mask", "acc"])


df.head()
```

```python
df=df[df['mask'].isin([50, 100])]
df16=df[df['patchSize'] == 16 ].copy()
df32=df[df['patchSize'] == 32].copy()
```

```python
import pandas as pd

df_sorted = df.sort_values(by=[ 'patchSize', 'topK', 'mask',  'depth',], ascending=[True, True, True, True])

df_sorted.head()
```

```python
import matplotlib.pyplot as plt
import numpy as np

groups = df_sorted.groupby(['patchSize', 'topK'])

for (patchSize, topK), group in groups:
    plt.figure(figsize=(8, 6))
    
    min_val = group['acc'].min()
    max_val = group['acc'].max()
    value_range = max_val - min_val

    margin = value_range * 0.1 
    adjusted_min = min_val - margin
    adjusted_max = max_val + margin
    
    for mask, sub_group in group.groupby('mask'):
        sub_group = sub_group.sort_values(by='depth') 
        plt.plot(
            sub_group['depth'], 
            sub_group['acc'], 
            marker='o', 
            linestyle='-', 
            linewidth=2, 
            label=f'Mask: {mask}'
        )
    
    plt.ylim(adjusted_min, adjusted_max)  
    plt.xlabel('Depth')
    plt.ylabel('Accuracy')
    plt.title(f'PatchSize: {patchSize}, TopK: {topK}')
    plt.xticks(group['depth'].unique()) 
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.legend(title="Mask")  
    
    plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np  # Add NumPy for calculating the mean
from matplotlib.ticker import FuncFormatter

sns.set_style("whitegrid")
topkL=[1,5,10,20,50]
depthL=[3,6,9,12]
maskL=[10,30,50,100]
prune_data1 =[[[loaded_data[f'vit-b-16-top{t}-d{d}-p{m}'][1] for t in topkL] for m in maskL] for d in depthL]
box_colors = ['skyblue', 'lightgreen', 'lightcoral', '#f6c660']  # Removed lightpink for baseline
# box_colors = ['#D6CCE5', '#90EE90', '#D2E9CF', '#f6c660']  # skyblue, lightgreen, lightcoral, yellow in HEX

def percentage_formatter(x, pos):
    return '{:.0f}%'.format(x * 100)
# Extract baseline value
baseline_value = loaded_data["baseline_vit_b_16"][0].mean()
labels = ["mask 10%", "mask 30%", "mask 50%", "mask 100%"]
x_labels = [f"top{top}" for top in topkL]


for patch in [16,32]:
    for i,d in enumerate(depthL):
        # Name = f"Mean Acc of Enhance To ViT-{patch}\n(depth {d})"
        Name = f"Mean Acc of Prune To ViT-{patch}\n(depth {d})"
        # Remove baseline from prune data
        prune_data =prune_data1[i]

        # Create a figure and axis objects
        fig, axes = plt.subplots(nrows=1, ncols=4, figsize=(15, 3))
        # Adjust font sizes
        plt.rcParams.update({'font.size': 10})

        # Plot the data using barplot
        for subplot_id, (model_data, label, ax) in enumerate(zip(prune_data, labels, axes)):
            # Convert list of data into a suitable format for sns.barplot
            sns.barplot(data=model_data, ax=ax, palette=box_colors, estimator=np.mean, errorbar=None)

            # Add baseline as a horizontal dashed line
            ax.axhline(y=baseline_value, color='grey', linestyle='--', linewidth=1.5, label='baseline')

            ax.set_title(label, fontsize=12)
            if not subplot_id: 
                ax.set_ylabel(Name, fontsize=12)
            ax.set_xticklabels(x_labels, fontsize=12)
            ax.tick_params(axis='both', which='major', labelsize=12)
            ax.grid(True, linestyle='--', alpha=0.5)  # Add grid lines
            # ax.set_ylim(0.0, 0.1)
            ax.set_ylim(0.65, 0.75)
            
            # Set y-axis formatter to percentage
            ax.yaxis.set_major_formatter(FuncFormatter(percentage_formatter))

            # Add legend for the baseline
            ax.legend(loc='upper right', fontsize=10)

        # Adjust spacing between subplots
        plt.subplots_adjust(wspace=0.25)

        # Save the figure
        Nname=Name.replace('\n','-')
        plt.savefig(f"imgs/MeanAcc_{Nname}.pdf", bbox_inches="tight")
```
