```python
import os
import jsonlines
import numpy as np

def load_results_jsonl(path):
    with jsonlines.open(path) as reader:
        score = []
        _data = [obj for obj in reader]
        for _ in _data:
            for one_data in _:
                score.append(one_data['ja_score'])
        score = np.array(score).squeeze(1)

    return score

def load_results_jsonl_ja(path):
    with jsonlines.open(path) as reader:
        score = []
        _data = [obj for obj in reader]
        for _ in _data:
            for one_data in _:
                score.append(one_data['ja_score'])
        score = np.array(score).squeeze(1).squeeze(1)

    return score

def load_results_jsonl_base(path):
    with jsonlines.open(path) as reader:
        score = []
        _data = [obj for obj in reader]
        for _ in _data:
            for one_data in _:
                score.append(one_data['base_score'])
        score = np.array(score).squeeze(1).squeeze(1)

    return score

    
```

# VIT-B-16

```python
JA_PATH = 'PATH_TO_JOINT_ATTRIBUTION_RESULTS'
INFLUENCE_PATTERN_PATH = 'PATH_TO_INFLUENCE_PATTERN_RESULTS'
BASE_PATH = 'PATH_TO_BASE_RESULTS'

integrated_gradients_scores = []
influence_pattern_scores = []
ja_scores = []
base_scores = []

for class_idx in range(1000):
    influence_pattern_path = os.path.join(INFLUENCE_PATTERN_PATH, f"imgnet_all-{int(class_idx)}.rlt.jsonl")
    ja_path = base_path = os.path.join(JA_PATH, f"imgnet_all-{int(class_idx)}.rlt.jsonl")

    influence_pattern_score = load_results_jsonl(influence_pattern_path)
    ja_score = load_results_jsonl_ja(ja_path)
    base_score = load_results_jsonl_base(base_path)

    influence_pattern_scores.append(influence_pattern_score)
    ja_scores.append(ja_score)
    base_scores.append(base_score)

mean_influence_pattern_score_vit_B_16 = np.array(influence_pattern_scores).reshape(-1, 50).mean(1)
mean_ja_score_vit_B_16 = np.array(ja_scores).reshape(-1, 50).mean(1)
mean_base_score_vit_B_16 = np.array(base_scores).reshape(-1, 50).mean(1)

activation_data_vit_B_16 = mean_base_score_vit_B_16
influence_pattern_data_vit_B_16 = mean_influence_pattern_score_vit_B_16
ours_data_vit_B_16 = mean_ja_score_vit_B_16

vit_b_32_data = [
    activation_data_vit_B_16,
    influence_pattern_data_vit_B_16,
    ours_data_vit_B_16,
]
```
